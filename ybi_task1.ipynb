# **case 1**

```python
print("churn,when the monthly charges are greater than mean pf monthly charges")
df.groupby(a)["Churn"].value_counts()
```

```python
df.groupby(b)["Churn"].value_counts()
```

```python
df.groupby([a,b])["Churn"].value_counts()
```

```python
# Create the groupby object
grouped_data = df.groupby([a, b])["Churn"].value_counts(normalize=True).unstack()

grouped_data.plot(kind='bar', figsize=(10, 6))
plt.title('Churn Rate by Monthly and Total Charges')
plt.xlabel('Monthly and Total Charges above average (True/False)')
plt.ylabel('Churn Rate')
plt.xticks(rotation=0)
plt.legend(title='Churn')
plt.tight_layout()
plt.show()
```
